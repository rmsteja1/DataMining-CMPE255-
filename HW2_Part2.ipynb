[LP] Perform Association Rule Mining using the apriori algorithm.
1. If you are using the February dataset, run the algorithm with a support above 0.55 and confidence above 0.88.
2. If you are using the March dataset, run the algorithm with a support above 0.6 and confidence above 0.8.

The library you can use is mlxtend. The imports are included below. 



```python
## Your code for LP goes here.
import mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

## Selecting support value greater than 0.6
march_frequent_itemsets=apriori(data_frame_march,min_support=0.6,use_colnames=True)
march_frequent_itemsets

```

```python
## Selecting rules having confidence greater than 0.8
march_result=association_rules(march_frequent_itemsets,metric="confidence",min_threshold=0.8)
march_result
```

From the above results we can observe the following insights:
1)For the association rule "Market St at 10th St", "Powell St BART Station (Market St at 4th St)" the support is 60% and the confidence is 84%.

2)For the association rule "Market St at 10th St", "Powell St BART Station (Market St at 4th St)" the support is 60% and the confidence is 89%.

Association Rule:
1) Association rules can be compared to simple if/then statements that helps in finding relations among seemingly independent data lables.
2) Association rule works on non-categorical data sets.
3) Association rule has 2 parts, antecedent and consequent.
4) Anticedent is something occured in the data where consequent is found in combination of Anticedent

On basis of following paramaters, the above relation ships are observed.
i) Support: It defines the number of times if/then relation appears in the database.
ii) Confidence: Confidence tells the number of times these relations have found to be true.

Hence, association rule mining tries to observe the rules that govern how or why such products/items are frequently bought together.
